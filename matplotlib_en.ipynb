# Python Reference-Guide - Matplotlib

### import Matplotlib

```python
import matplotlib.pyplot as plt
```

### Create a Plot from Numpy array

```python
import numpy as np
numpy_data = np.array([[1,2,3,4,5], [1,4,9,16,25]])
```

```python
plt.plot(numpy_data[0],numpy_data[1])
plt.show
```

### Labeling a Plot

```python
plt.plot(numpy_data[0],numpy_data[1])
plt.title("Plot Title")
plt.xlabel("x - label")
plt.ylabel("y - label")
plt.show
```

### Figuresize

```python
plt.figure(figsize=(15,5))
plt.plot(numpy_data[0],numpy_data[1])
plt.show
```

## Plottypes

### Bars

```python
divisions =["A","B","C","D"]
divisions_marks =[60,75,90,83]
plt.bar(divisions,divisions_marks)
plt.show()
```

```python
divisions =["A","B","C","D"]
divisions_marks =[60,75,90,83]
plt.barh(divisions,divisions_marks)
plt.show()
```

### Pie

```python
objects = ["Red","Blue","Green","Yellow"]
share = [23.4,5.3,12,43]
explode = (0, 0.1, 0, 0)
plt.pie(share, explode=explode, labels=objects, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
```

### Histogram

```python
x = np.random.randn(1000)
```

```python
plt.hist(x, 100)
plt.show()
```

### Scatter Plot

```python
height = np.array([153,176,201,154])
weight = np.array([63,78,90,32])
plt.scatter(height,weight)
plt.show()
```
